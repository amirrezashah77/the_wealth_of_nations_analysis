```python
# Import our classes from the main file
from main import DataManager, Analyzer, ChartBuilder
import matplotlib.pyplot as plt

# This magic command makes charts appear directly in the notebook
%matplotlib inline 

# 1. Load Data
manager = DataManager()
df = manager.get_clean_data()
print(f"Data Loaded: {len(df)} rows")

# 2. Test Regression (1990)
data_1990 = df[df['Year'] == 1990]
analyzer = Analyzer(data_1990)
slope, intercept = analyzer.run_regression()
r2 = analyzer.get_r_squared()

print(f"R-Squared for 1990: {r2:.3f}")

# 3. Show Scatter Chart
ChartBuilder.plot_scatter(data_1990, slope, intercept)

fig2 = ChartBuilder.plot_time_series(df, "United States")
plt.show()
```
